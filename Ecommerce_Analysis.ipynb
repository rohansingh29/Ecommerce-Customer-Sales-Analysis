```python
df = df.merge(df_payments, on='order_id', how='left')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['product_category_name'] = df['product_category_name'].fillna('unknown')

for col in ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

CREATING USEFUL COLUMNS


```python
# Delivery time (days between purchase and delivery)
df['delivery_time_days']=(df['order_delivered_timestamp']-df['order_purchase_timestamp']).dt.days
```

```python
#Delivery delay (negative = early, positive = late)
df['delivery_delay_days'] = (df['order_delivered_timestamp'] - df['order_estimated_delivery_date']).dt.days
```

```python
#Order month/year
df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_month'] = df['order_purchase_timestamp'].dt.month
```

```python
#Total order value (price + shipping)
df['total_order_value'] = df['price'] + df['shipping_charges']
```

```python
df[['delivery_time_days', 'delivery_delay_days', 'order_year', 'order_month', 'total_order_value']].describe()
```

```python
df.sort_values('delivery_delay_days', ascending=False).head()
```

EXPLORATORY DATA ANALYSIS


```python
#Revenue By Month
df.groupby(['order_year', 'order_month'])['total_order_value'].sum().reset_index()
```

Revenue grew steadily from late 2016 through 2018, peaking around November 2017 (₹31.5L) and stabilizing at ₹20-26L/month through mid-2018. September 2018 shows only ₹235 due to partial data collection at the end of the dataset.

```python
#Top 10 product categories by revenue
df.groupby('product_category_name')['total_order_value'].sum().sort_values(ascending=False).head(10)
```

75% of orders are labeled under the "toys" category, an implausible concentration compared to all other categories combined. This is likely a placeholder/default value in the source data rather than an accurate label.


```python
# Payment method distribution
df['payment_type'].value_counts(normalize=True) * 100
```

Credit card dominates at ~74% of all payments 

Wallet is a distant second at ~19%

Voucher (5.5%) and debit card (1.4%) are minor payment methods

```python
# Delivery performance 
df['delivery_delay_days'].apply(lambda x: 'Late' if x > 0 else 'On-time/Early')
df['delivery_status'].value_counts(normalize=True) * 100
```

Approx 94% delivery is on time and 6% are late it means the delivery performance is very good.

```python
# Revenue/orders by state
df.groupby('customer_state')['total_order_value'].sum().sort_values(ascending=False).head(10)
```

Sao Paulo (SP) accounts for the largest share of revenue (~₹1.46 crore), consistent with it having the highest customer base (37,879 customers). RJ and MG follow as the next largest markets.
